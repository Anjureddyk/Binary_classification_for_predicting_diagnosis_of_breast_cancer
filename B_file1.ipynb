```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data = pd.read_csv("breast-cancer.csv")
```

```python
data.head()
```

```python
data.describe()
```

```python
data.isnull().sum()
```

```python
sns.pairplot(data, hue='diagnosis', vars=['radius_mean', 'texture_mean', 'perimeter_mean', 'area_mean'])

# Correlation matrix heatmap
corr_matrix = data.corr()
sns.heatmap(corr_matrix, cmap='coolwarm', annot=False)

# Box plot
sns.boxplot(x='diagnosis', y='radius_mean', data=data)

plt.show()
```

```python
# Generate a correlation matrix heatmap
corr = data.corr()
plt.figure(figsize=(20,20))
sns.heatmap(corr, annot=True, cmap='coolwarm')

# Plot a histogram of the diagnosis variable
sns.histplot(data=data, x='diagnosis')

# Plot a scatterplot of radius_mean vs. area_mean
sns.scatterplot(data=data, x='radius_mean', y='area_mean', hue='diagnosis')
```

```python
# Step 2: Drop the 'id' column and save the modified DataFrame
data.drop('id', axis=1, inplace=True)

# Step 3: Save the pre-processed data to a new CSV file
data.to_csv("pre_processed_data.csv", index=False)
```
